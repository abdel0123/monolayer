```python
import matplotlib.pyplot as plt

def ET(a,b):
    if a == 1 and b == 1:
        return 1
    return 0

def OU(a,b):
    if a == 0 and b == 0:
        return 0
    return 1

def XOR(a,b):
    if a == b:
        return 0
    return 1

def appren(f_choisie):
    w1 = w2 = b = ConteurIter = 0
    alpha = 0.7
    E = []
    # Le tableau d'entrée
    x = [[0,0],[0,1],[1,0],[1,1]]
    # on va répéter les iteration jusqu'a ce que la fonction appris. Dans notre cas le max d'iteration est 20
    for k in range(20):
        s = 0
        # on va calculer les nouvelles valeurs d'entrer w1,w2 pour chaque élément du vecteur x
        for i in range(4):
            a = w1*x[i][0]+w2*x[i][1]+b
            if a < 0:
                y = 0
            else:
                y = 1
            # on va definir la valeur de d selon la fonction choisie par l'utilisateur
            if f_choisie == 1:
                d = ET(x[i][0],x[i][1])
            elif f_choisie == 2:
                d = OU(x[i][0],x[i][1])
            elif f_choisie == 3:
                d = XOR(x[i][0],x[i][1])
            else:
                break
            e = d-y
            w1 = w1 + alpha*e*x[i][0]
            w2 = w2 + alpha*e*x[i][1]
            b = b + alpha*e*1
            s = s + e*e
      
        E.append(s/2)
       
        if E[-1] > 0:
            ConteurIter = ConteurIter + 1

            
    return w1,w2,b,ConteurIter,E
# on va tester la fonction
while(1):
    print(" Apprendre par le perceptron monocouche")
    print("ET ---- 1")
    print("OU------2")
    print("XOR-----3")
    print("Sortir--0")
    print("-------------------------------------------------------------------------------------------------")
    inp = input("choisir la fonction logique : ")
    inp2 = int(inp)
    if inp2 == 1:
        print(" la fonction logique  ET ")
    elif inp2 == 2:
        print(" la fonction logique  OU ")
    elif inp2 == 3:
        print(" la fonction logique  XOR")
    else:
        print("fin")
        break
    
    w1,w2,b,c,E = appren(inp2)
    print("Entrainement termine")
    print("w1 = ",w1)
    print("w2 = ",w2)
    print("b = ",b)
    print("la fonction a appris dans iteration numero : ",c)
    plt.plot(E)
    plt.title("la courbe EQM en fonction du NI")
    plt.show()
    print("LES TEST")
    choix2 = ''
    while choix2 != 'n':
        inp = input("x1 : ")
        x1 = int(inp)
        while x1 != 1 and x1 != 0:
            inp = input("x1 doit etre 0 ou 1, veulliez saisir x1 : ")
            x1 = int(inp)
        inp = input("saisie le x2 : ")
        x2 = int(inp)
        while x2 != 1 and x2 != 0:
            inp = input("x2 doit etre 0 ou 1, veulliez saisir x2 :")
            x2 = int(inp)
        if(x1*w1+x2*w2+b < 0):
            print("le resultat ----> 0")
        else:
            print("le resultat ----> 1")
        choix2 = input("tester la fonction o/n : ")
        
        
    



```
